```python
import pandas as pd
import numpy as np
```

```python
def fetch_data_df():
    df = pd.read_csv('../../data/asset_classes.csv', index_col=0)
    df.index = pd.to_datetime(df.index)
    df = df.iloc[:, :5]
    return df
```

```python
df = fetch_data_df()
```

```python
returns_df = df.pct_change(fill_method=None).dropna()
returns_df
```

```python
weights = np.ones(5) / 5
portfolio_returns = np.dot(returns_df, weights)
```
