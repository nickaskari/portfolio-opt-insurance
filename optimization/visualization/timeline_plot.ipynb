```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../data/final_asset_classes.csv', index_col=0)
df.index = pd.to_datetime(df.index)
```

```python
scenario1 = ("2021-11-01", "2022-11-01")
scenario2 = ("2007-06-01", "2008-06-01")
scenario3 = ("2022-11-01", "2023-11-01")
```

```python
scenarios = [scenario1, scenario2, scenario3]
colors = ['lightblue', 'lightgreen', 'lightpink']  # Colors for the scenarios

# Plot the data
plt.figure(figsize=(12, 6))

# Highlight each scenario
for i, scenario in enumerate(scenarios):
    plt.axvspan(pd.to_datetime(scenario[0]), pd.to_datetime(scenario[1]), 
                color=colors[i], alpha=0.4, label=f'Scenario {i+1}')

# Plot MSCIWORLD returns
plt.plot(df.index, df['MSCIWORLD'], label='MSCIWORLD', color='blue', linewidth=1.5)

# Add labels, legend, and grid
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.title('MSCIWORLD with Highlighted Scenarios')
plt.legend(loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
