```python
#calculate daily return of SPY, remember that top row is the latest date

new_df = df[['Exchange Date', 'SPY.US.Equities']]

new_df['Percentage Gain'] = df['SPY.US.Equities'].iloc[::-1].pct_change(fill_method=None) * 100

new_df.head()
```

```python
#plot the distribution of daily return for SPY

px.histogram(new_df, x='Percentage Gain').show()
```

```python
# find the date of the highest and lowest daily return

highest_return = new_df['Percentage Gain'].max()

highest_return_date = new_df[new_df['Percentage Gain'] == highest_return]['Exchange Date']

lowest_return = new_df['Percentage Gain'].min()

lowest_return_date = new_df[new_df['Percentage Gain'] == lowest_return]['Exchange Date']

print(f'The highest daily return for SPY is {highest_return:.2f}% on {highest_return_date}')

print(
    f'The lowest daily return for SPY is {lowest_return:.2f}% on {lowest_return_date}')
```
